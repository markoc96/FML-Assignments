# Introduction

I have selected the Fifa 18 dataset because I have a strong knowledge of this topic. I will be analysing this data and then creating a model to predict the playing position of the players in the dataset. The dataset contains ~75 variables relating to ~17,000 FIFA players. These variables can be categorised as Personal, Attribute or Playing Position data. 

### Selecting the variables

Based on my domain knowledge, I know that the personal data will have no correlation with the playing position and so I will exclude this from my analysis. On inspection of the data, I have decided to only use the attribute data to make the exercise more manageable, but I acknowledge that including the playing position better model could create a better model.

```python
import seaborn as sns
import sklearn as sk
import pandas as pd
import matplotlib.pyplot as plt
import math
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
import numpy as np

df = pd.read_csv("CompleteDataset.csv")
df.info()
```

```python
#dropping all columns that won't be used
#personal data
df = df.drop(['Name', 'Age', 'Club', 'Club Logo', 'Photo', 'Nationality', 'Flag', 'ID'], axis=1)
df.info()
```

```python
#playing position data - I tried using this in previous iterations and it didn't really work well
df.drop(df.iloc[:, 40:55], inplace = True, axis = 1)
df.info()
```

```python
df.drop(df.iloc[:, 41:52], inplace = True, axis = 1)
df.info()
```

```python
df.drop(df.columns[0], inplace = True, axis = 1)
df.info()
```

```python
#remove euro sign and M/K from wage and value columns

df['Wage'] = df['Wage'].str[1:]
df['Value'] = df['Value'].str[1:]

df['Wage'] = df['Wage'].str[:-1]
df['Value'] = df['Value'].str[:-1]

#remove null values
df = df[df['Value'] != '']
df = df[df['Wage'] != '']

#convert to float/int for analysis
df['Value'] = df['Value'].astype(np.float64)
df['Wage'] = df['Wage'].astype(int)

df.head()
```

```python
#splitting the preferred positions column
df['Preferred Positions'] = [x.split(' ')[-0] for x in df['Preferred Positions']]
df.head()
```

```python
#creating categorical codes for the positions
df['Preferred Positions'] = pd.Categorical(df['Preferred Positions'])
df['code'] = df['Preferred Positions'].cat.codes
```

```python
#making all the columns numeric as there were some special characters in some cells
cols = ['Acceleration', 'Aggression', 'Agility', 'Balance', 'Ball control', 'Composure', 'Crossing', 'Curve', 
'Dribbling', 'Finishing', 'Free kick accuracy', 'GK diving', 'GK handling', 'GK kicking', 'GK positioning', 'GK reflexes',
'Heading accuracy', 'Interceptions', 'Jumping', 'Long passing', 'Long shots', 'Marking', 'Penalties', 'Positioning', 'Reactions',
'Short passing', 'Shot power', 'Sliding tackle','Sprint speed', 'Stamina', 'Standing tackle', 'Strength', 'Vision', 'Volleys']

df[cols] = df[cols].apply(pd.to_numeric, errors='coerce', axis=1)
```

```python
df.info()
```

```python
from sklearn.model_selection import train_test_split

df = df.dropna()

x = df[cols]
y = df['code']

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.3, random_state=1)
```

```python
from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier(n_neighbors=3)
knn = knn.fit(x_train, y_train)
```

```python
knn.score(x_test,y_test)
```
